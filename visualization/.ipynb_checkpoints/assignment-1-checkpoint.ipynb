#### Transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
columns_ = iris.feature_names
columns_
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris_data = pd.DataFrame(iris.data, columns = columns_)
```

```python
iris_data['type'] = iris.target
```

```python
def changeTypNumberToName(number_):
    if number_ == 0:
        return 'setosa'
    elif number_ == 1:
        return 'versicolor'
    return 'virginica'
iris_data.type = iris_data.type.apply(changeTypNumberToName)
```

```python
iris_data
```

```python
fig = plt.figure(1, figsize=(10, 6))
axes3D = Axes3D(fig, elev=-150)
axes3D.scatter(iris_data['sepal length (cm)'], iris_data['sepal width (cm)'], c=iris.target, s=50)
axes3D.set_xlabel("setosa")
axes3D.set_ylabel("versicolor")
axes3D.set_zlabel("virginica")
```

```python
fig = plt.figure(1, figsize=(10, 6))
axes3D = Axes3D(fig, elev=-150)
axes3D.scatter(iris_data['sepal length (cm)'], iris_data['petal length (cm)'], c=iris.target, s=50)
axes3D.set_xlabel("setosa")
axes3D.set_ylabel("versicolor")
axes3D.set_zlabel("virginica")
plt.show()
```

```python
fig = plt.figure(1, figsize=(10, 6))
axes3D = Axes3D(fig, elev=-150)
axes3D.scatter(iris_data['sepal length (cm)'], iris_data['petal width (cm)'], c=iris.target, s=50)
axes3D.set_xlabel("setosa")
axes3D.set_ylabel("versicolor")
axes3D.set_zlabel("virginica")
plt.show()
```

```python
fig = plt.figure(1, figsize=(10, 6))
axes3D = Axes3D(fig, elev=-150)
axes3D.scatter(iris_data['petal width (cm)'], iris_data['sepal length (cm)'], c=iris.target, s=50)
axes3D.set_xlabel("setosa")
axes3D.set_ylabel("versicolor")
axes3D.set_zlabel("virginica")
plt.show()
```
